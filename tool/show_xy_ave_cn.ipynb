```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

def show_ave_cn(data1,data2,dataset_name,axvline = 3):
    density = gaussian_kde(data1)
    x = np.linspace(min(data1), max(data1), 100)
    y = density(x)

    plt.plot(x, y,label = 'Non connected node pairs',c = 'red')

    density = gaussian_kde(data2)
    x = np.linspace(min(data2), max(data2), 100)
    y = density(x)

    plt.plot(x, y, label = 'Connected node pairs',c = 'green')
    plt.axvline(x=axvline, color='blue', linestyle='--',label = 'x = ' + str(axvline))
    plt.ylim(0,2)
    plt.xlabel('Average common nodes')
    plt.ylabel('Probability Density')
    plt.title(dataset_name)

    plt.legend()

    plt.savefig("../result/xy/ave_cn_" + dataset_name +  ".png", dpi=300)
    plt.savefig("../result/xy/ave_cn" + dataset_name +  ".pdf")

    plt.savefig("../result/xy/ave_cn" + dataset_name +  ".eps")
    plt.show()
```

```python
acvlines = [3,2.7,2.6,3.2,2.9,3.2]
datasets = ['S. pombe','C. elegans','S. cerevisiae','A. thaliana','D. Melanogaster','HuRI']
datasets_name = ['S. pombe','C. elegans','S. cerevisiae','A. thaliana','D. melanogaster','H. sapiens']
for i in range(6):
    x1 = []
    x2 = []
    dataset = datasets[i]
    filename = 'D:/code/edit/c++/SMS_new/temp/' + dataset + '_cn_and_edges_density.txt'  # 文件名
    # 按行读取数据
    with open(filename, 'r') as file:
        lines = file.readlines()
        for line in lines:
            w = line.strip().split(",")
            # print(w)
            if len(w) != 3:
                continue
            if int(w[2]) == 0:
                x1.append(float(w[0]))
            else:
                x2.append(float(w[0]))

    show_ave_cn(x1,x2,datasets_name[i],acvlines[i])
```
